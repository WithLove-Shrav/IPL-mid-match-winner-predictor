```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline

# Assuming 'pipe' is a pre-trained model object with a 'predict_proba' method

def replace_inf(df, fill_value=0):
    df.replace([np.inf, -np.inf], np.nan, inplace=True)  # Replace inf/-inf with NaN
    df.fillna(fill_value, inplace=True)  # Fill NaN with the specified value
    return df

def check_inf(df):
    if df.isin([np.inf, -np.inf]).any().any():
        print("DataFrame contains infinite values.")
        df = replace_inf(df)  # Call the function to replace inf/-inf with NaN and fill NaN
    else:
        print("No infinite values found in DataFrame.")
    return df

def match_progression(x_df, id, pipe):
    # Filter for the specific match
    match = x_df[x_df['id'] == id]
    if match.empty:
        print(f"No data found for match ID: {id}")
        return None, None

    # Ensure the match DataFrame has all the required columns before proceeding
    required_columns = ['bowling_team', 'total_runs_x', 'city', 'batting_team']
    missing_columns = [col for col in required_columns if col not in match.columns]
    if missing_columns:
        print(f"Missing columns in the match DataFrame: {missing_columns}")
        return None, None

    # Sort by over and ball to ensure correct ordering
    match = match.sort_values(by=['over', 'ball'])
    
    # Find the last delivery of each over
    end_of_over = match.drop_duplicates(subset='over', keep='last')
    
    # Drop rows with any NaNs in specified columns
    temp_df = end_of_over.dropna(subset=['runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr'])
    temp_df = temp_df[temp_df['balls_left'] != 0]
    
    # Check if temp_df is empty before prediction
    if temp_df.empty:
        print("No data available for the given match ID after filtering.")
        return None, None
    
    # Preprocess the DataFrame to handle potential infinite values
    temp_df = check_inf(temp_df)
    
    # Predict probabilities
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0] * 100, 1)
    temp_df['win'] = np.round(result.T[1] * 100, 1)
    
    # Generate end_of_over sequence
    temp_df['end_of_over'] = range(1, temp_df.shape[0] + 1)
    
    # Calculate target
    target = temp_df['total_runs_x'].values[0]
    
    # Calculate runs_after_over
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0, target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    
    # Calculate wickets_in_over
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0, 10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    # Select relevant columns to return
    temp_df = temp_df[['end_of_over', 'runs_after_over', 'wickets_in_over', 'lose', 'win']]
    return temp_df, target

# Example usage of the match_progression function
# Assuming 'merged_deliverydf' is a DataFrame containing the match data
# and 'pipe' is your prediction pipeline
temp_df, target = match_progression(merged_deliverydf, 1254101, pipe)

# Print the result if available
if temp_df is not None:
    print(f"Target- {target}")
    print(temp_df)
else:
    print("No data to display.")
```
